```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
fake = pd.read_csv("dataset/Fake.csv")
true = pd.read_csv("dataset/True.csv")
```

```python
true['label'] = 1
```

```python
true.head()
```

```python
fake['label'] = 0
```

```python
fake.head()
```

```python
combined_news = pd.concat([fake, true], axis = 0, ignore_index = True )
```

```python
combined_news.drop(['title','subject','date'], axis=1,inplace=True)
```

```python
shuffled_news = combined_news.sample(frac = 1)
```

```python
shuffled_news.head()
```

```python
shuffled_news.reset_index(inplace = True)
```

```python
shuffled_news.drop(['index'], axis = 1, inplace = True)
```

```python
import re
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+','', text)
    text = re.sub(r'<.*?>','',text)
    text = re.sub(r'[^\w\s]','',text)
    text = re.sub(r'\d','',text)
    text = re.sub(r'\n',' ',text)
    return text
```

```python
shuffled_news['text'] = shuffled_news['text'].apply(wordopt)
shuffled_news.head()
```

```python
x, y = shuffled_news['text'], shuffled_news['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()
DTC.fit(xv_train, y_train)
predict_dtc = DTC.predict(xv_test)
DTC.score(xv_test, y_test )
```

```python
print(classification_report(y_test, predict_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(xv_train, y_train)
```

```python
predict_rfc = rfc.predict(xv_test)
rfc.score(xv_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
gbc.fit(xv_train, y_train)
```

```python
predict_gbc = gbc.predict(xv_test)
rfc.score(xv_test, y_test)
```

```python
def result(x):
    if x == 0:
        return 'These news are potentially fake.'
    elif x == 1:
        return 'These news are most probably geniune.'
```

```python
def model(news):
    testing_news = {'text' : [news] }
    model_test = pd.DataFrame(testing_news)
    model_test['text'] = model_test['text'].apply(wordopt)
    model_x_test = model_test['text']
    model_xv_test = vectorization.transform(model_x_test)
    predict_lr, predict_dtc, predict_rfc, predict_gbc = LR.predict(model_xv_test),DTC.predict(model_xv_test),\
    rfc.predict(model_xv_test),gbc.predict(model_xv_test)
    return 'LR Prediction: {} \nGBC Prediction: {} \nDTC Prediction: {} \nRFC Prediction: {}'.format(
        result(predict_lr[0]), result(predict_gbc[0]), result(predict_dtc[0]), result(predict_rfc[0]))
```

```python
import joblib

LR = joblib.load('logistic_regression_model.pkl')
DTC = joblib.load('decision_tree_model.pkl')
rfc = joblib.load('random_forest_model.pkl')
gbc = joblib.load('gradient_boosting_model.pkl')
vectorization = joblib.load('tfidf_vectorizer.pkl')
```
